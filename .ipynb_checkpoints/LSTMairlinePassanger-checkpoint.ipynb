# LSTM on Airline passenger with LSTM keras

```python
import math 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("international-airline-passengers.csv" , usecols=[1] , skipfooter = 3)
```

```python
df.head(3)
```

# visualize

```python
plt.figure(figsize = (15,5))
plt.plot(df, label= "Airline passengers")
plt.xlabel("Months")
plt.ylabel("1000 international air passengers")
plt.title("airline passengers 1949-1960")
plt.legend()
plt.show()
```

# Data Preparation

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data_raw = df.values.astype("float32")

## normalize data in the (0,1) 
scaler = MinMaxScaler(feature_range =(0,1))
dataset = scaler.fit_transform(data_raw)
```

```python
Train_Size = 0.75

train_size = int(len(dataset)* Train_Size)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size, :] , dataset[train_size:len(dataset) , :]
```

```python
train.shape
```

```python
test.shape
```

# Get Data int shape to use in Keras

```python
def create_dataset(dataset, window_size = 1):
    data_X, data_Y = [] , []
    for i in range(len(dataset) - window_size -1):
        a = dataset[i:(i+window_size), 0 ]
        data_X.append(a)
        data_Y.append(dataset[i+window_size,0])
    return(np.array(data_X), np.array(data_Y))
```

```python
window_size = 1
train_X , train_Y = create_dataset(train, window_size)
test_X , test_Y = create_dataset(test, window_size)
```

# Reshape the data for keras 

```python
train_X = np.reshape(train_X , (train_X.shape[0] , 1 , train_X.shape[1]))
test_X = np.reshape(test_X , (test_X.shape[0] , 1 , test_X.shape[1]))
print(train_X.shape)
```

```python
print(test_X.shape)
```

# LSTM    input layer --> 4 Neuron layer  --> one dense layer to produce  and use MSE function for loss function

```python
from keras.models  import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()

model.add(LSTM(4, input_shape=(1, window_size)))
model.add(Dense(1))

### compile 

model.compile(loss ="mean_squared_error" , 
             optimizer="adam")
```

```python
model.fit( train_X , train_Y, epochs =100, batch_size =1 , verbose = 2)
```
