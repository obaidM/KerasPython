# DEEP NEURAL NETWORK

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np 
```

# Load the Dataset pima indians diabetes

```python
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
X= dataset[:,0:8]  ## predictors
Y=dataset[:,8]      ## response , target variable
```

# Add multiple layers

```python
model= Sequential()

model.add(Dense(12, input_dim=8,activation='relu'))
## the first layer is input_dim with 8 features and goes into 13 precepron layer
model.add(Dense(8,activation='relu'))
## secodn layer is 8 preceptron layer. This is a hidden layer
model.add(Dense(1, activation='sigmoid'))

# the output layer that only has one preceptron with sigmoid as this is your predicting layer
```

# Compile the layer

```python
model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])
```

# Fit the model

```python
model.fit(X,Y,epochs=150,batch_size=10)
```

```python
scores= model.evaluate(X,Y)
print("\n%s: %.2f%%" % (model.metrics_names[1],scores[1]*100))
```
